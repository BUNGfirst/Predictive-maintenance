```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
from sklearn.utils import shuffle
```

```python
raw_data = pd.read_csv('predictive_maintenance.csv')
raw_data
```

```python
data = raw_data.copy()
data = shuffle(data, random_state=20)
data
```

```python
df_class1 = data[data['Target'] == 1]
df_class2 = data[data['Target'] == 0].sample(n=len(df_class1), random_state=42)
```

```python
data_balanced = pd.concat([df_class1, df_class2])
data_balanced = shuffle(data_balanced, random_state=25)
data_balanced
```

```python
data_balanced = data_balanced.drop(['UDI', 'Product ID'], axis=1)
data_balanced
```

```python
data_balanced_dummies = pd.get_dummies(data_balanced, columns=['Type'])
data_balanced_dummies = data_balanced_dummies[['Type_H', 'Type_L', 'Type_M', 'Air temperature [K]', 'Process temperature [K]',
       'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]',
       'Target', 'Failure Type']]
data_balanced_dummies
```

```python
data_input = data_balanced_dummies[['Type_H', 'Type_L', 'Type_M', 'Air temperature [K]', 'Process temperature [K]',
       'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']]
data_input
```

```python
data_target1 = data_balanced_dummies['Target']
data_target1 = pd.DataFrame(data_target1, columns=['Target'])
data_target1 = data_target1.reset_index(drop=True)
data_target1
```

```python
column_to_scale = ['Air temperature [K]', 'Process temperature [K]',
                   'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']

scaled_array = preprocessing.scale(data_input[column_to_scale])
data_input_scaled = pd.DataFrame(scaled_array, columns=column_to_scale)

type_dummies = data_input[['Type_H', 'Type_L', 'Type_M']].reset_index(drop=True)

data_input_scaled = pd.concat([data_input_scaled.reset_index(drop=True), type_dummies], axis=1)
data_input_scaled = data_input_scaled[['Type_H', 'Type_L', 'Type_M', 'Air temperature [K]', 'Process temperature [K]',
       'Rotational speed [rpm]', 'Torque [Nm]', 'Tool wear [min]']]
data_input_scaled
```

```python
n = data_balanced.shape[0]
train_size = int(0.8*n)
validation_size = int(0.1*n)
test_size = n - train_size - validation_size

train_input = data_input_scaled[:train_size]
validation_input = data_input_scaled[train_size:train_size+validation_size]
test_input = data_input_scaled[train_size+validation_size:]

train_target = data_target1[:train_size]
validation_target = data_target1[train_size:train_size+validation_size]
test_target = data_target1[train_size+validation_size:]
```

```python
np.savez('predictive_train.npz', inputs=train_input, targets=train_target)
np.savez('predictive_validation.npz', inputs=validation_input, targets=validation_target)
np.savez('predictive_test.npz', inputs=test_input, targets=test_target)
```
